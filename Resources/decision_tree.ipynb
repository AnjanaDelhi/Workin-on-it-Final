```python
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
import os
```

```python
df = pd.read_csv(os.path.join("..", "Resources", "movies_complete_adjusted2.csv"))
df.head()
```

```python
# Adjusting for inflation
df['budget_1'] = df['budget'] * df['Normalize']
df['income_1'] = df['usa_gross_income'] * df['Normalize']
df.head()
```

```python
conditions = [(df["budget_1"] < 10000000), (df["budget_1"] >= 10000000) & (df["budget_1"] < 60000000), (df["budget_1"] >= 60000000)]
values = ['budget_low', "budget_med", "budget_high"]
df["budget_classes"] = np.select(conditions, values)
df.value_counts("budget_classes")
```

```python
conditions = [(df["income_1"] < 10000000), (df["income_1"] >= 10000000) & (df["income_1"] < 60000000), (df["income_1"] >= 60000000)]
values = ['income_low', "income_med", "income_high"]
df["income_classes"] = np.select(conditions, values)
df.value_counts("income_classes")
```

```python
df_1 = pd.concat([df,df.genre.str.get_dummies(sep=', ')],1)
df_1.head()
```

```python
#List the columns
list(df_1)

#df_1.to_excel("output.xlsx")  
```

```python
# Working on language, the first one is the primary language, hence we are keeping only the first one for simplicity
# splitting at occurrence of whitespace 
df_1["lang"]= df_1["language"].str.split(",").str[0] 

# displaying first element from list 
#df_1["language"].str.get(0) 
df_1.head()
 
```

```python
#Count movies in each of the languages so that we can choose a few languages
df_1.groupby(['lang']).size()
```

```python
# Include popular language movies only
df_2 = df_1.loc[df_1['lang'].isin(['English','French', 'Japanese', 'Mandarin', 'Spanish', 'Korean'])]
df_2.head()
```

```python
data = df_2[[ "Action", 'Adventure', 'Animation', 'Biography', 'Comedy', 'Crime', 'Drama', 'Family', 'Fantasy', 'Horror', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller',
 "lang", "budget_classes", 'income_classes']]
X = data.drop(['income_classes'], axis=1)
y = data["income_classes"].values.reshape(-1, 1)
print(X.shape, y.shape)
```

```python
from sklearn.preprocessing import LabelEncoder

# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y)
encoded_y = label_encoder.transform(y)
```

```python
# Dummy encoding
X = pd.get_dummies(X)
X.head()
```

```python
list(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
X_train.head()
```

```python
#Fitting the Standard scaler model to the training data
from sklearn.preprocessing import StandardScaler
X_scaler = StandardScaler().fit(X_train)
#y_scaler = StandardScaler().fit(y_train)
```

```python
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
#y_train_scaled = y_scaler.transform(y_train)
#y_test_scaled = y_scaler.transform(y_test)
```

```python
# Decision tree Model
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators= 750, n_jobs = 10000, verbose = 2)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```
